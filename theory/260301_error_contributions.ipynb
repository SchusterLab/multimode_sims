# RAM Surface Code Scaling with coherence time 
Consider gate time, and mode coherence

error correction cycle length \(t_{log}\). main error
error on physical qubit $$p = \frac{t_{log}}{T_{\text{coh},qub}}$$
error on memory $$p_{mem,idle} =  \frac{(N-1) *( t_{log} + 2t_{swap})}{T_{\text{coh},mem}}  $$
error on memory swap $$p_{mem, gate} = t_{swap}*(\frac{1}{T_{\text{coh}, mem}} + \frac{1}{T_{\text{coh}, qub}})$$


$$$$

x axis: t1 qub

at error rate 1e-10

## Remember fundamental equations 
### memory less system 
$$a = \left(\frac{p  }{p_{\text{threshold}}}\right)$$
$$\varepsilon = a^{\frac{D+1}{2}}$$
### memory full system 
$$b = \left(\frac{p_{mem, swap}}{p_{\text{mem threshold}}}\right)$$
$$c = \left(\frac{p_{mem, idle}}{p_{\text{mem threshold}}}\right)$$
$$\varepsilon' = (a+b+ cN)^{\frac{D'+1}{2}}$$

### Comparison of resources 
r = 3 (z + x + ancilla control)
Condition for memory efficiency
$$(r+1)D'^2 <rND^2$$
Define memory efficient 
$$eff = \frac{rND^2}{(r+1)D'^2 }$$

```python
import numpy as np
import matplotlib.pyplot as plt

def get_distance(vareps, base):
    """ D = 2 * log10(vareps) / log10(base) - 1 """
    if base >= 1.0 or base <= 0:
        return np.nan
    d = (2 * np.log10(vareps) / np.log10(base)) - 1
    return d if d > 0 else np.nan

def find_optimal_N_stats(vareps, a, b, c, r, N_max):
    D = get_distance(vareps, a)
    if np.isnan(D): 
        return np.nan, np.nan

    best_eff = -1
    n_opt = np.nan
    
    for n in range(1, N_max + 1):
        base_prime = a + b + c * n
        if base_prime >= 1.0: 
            break
            
        Dp = get_distance(vareps, base_prime)
        if np.isnan(Dp): 
            continue
            
        # Efficiency = (r * N * D^2) / ((r + 1) * D'^2)
        eff = (r * n * (D**2)) / ((r + 1) * (Dp**2))
        
        if eff > 1 and eff > best_eff:
            best_eff = eff
            n_opt = n
            
    return (best_eff, n_opt) if best_eff > 1 else (np.nan, np.nan)

def run_parameter_sweep(vareps=1e-10, a_fixed=0.1, r=2, N_max=100):
    b_vals = np.logspace(-3, 0, 100) #np.linspace(0.01, 0.5, 100)
    c_vals = np.logspace(-3, 0, 100) 
    B, C = np.meshgrid(b_vals, c_vals)
    
    eff_grid = np.full(B.shape, np.nan)
    n_opt_grid = np.full(B.shape, np.nan)

    for i in range(len(c_vals)):
        for j in range(len(b_vals)):
            eff, n_opt = find_optimal_N_stats(vareps, a_fixed, B[i,j], C[i,j], r, N_max)
            eff_grid[i,j] = eff
            n_opt_grid[i,j] = n_opt

    fig, ax = plt.subplots(1, 2, figsize=(15, 5))
    
    # Efficiency Plot - Using raw strings r'' to avoid SyntaxWarnings
    im1 = ax[0].pcolormesh(B, C, eff_grid, shading='auto', cmap='plasma', vmin=1.0, vmax=50)
    ax[0].set_yscale('log')
    ax[0].set_xscale('log')
    ax[0].set_title(fr'Max Efficiency ($a={a_fixed}$, $\epsilon={vareps}$)')
    ax[0].set_xlabel('Swap Error Contribution (b)')
    ax[0].set_ylabel('Idle Error per N (c)')
    cb1 = fig.colorbar(im1, ax=ax[0], extend='max')
    cb1.set_label('Efficiency Factor')

    # Optimal N Plot
    im2 = ax[1].pcolormesh(B, C, n_opt_grid, shading='auto', cmap='viridis', vmin=1, vmax=N_max)
    ax[1].set_yscale('log')
    ax[1].set_xscale('log')
    ax[1].set_title(fr'Optimal N ($a={a_fixed}$, $N_{{max}}={N_max}$)')
    ax[1].set_xlabel('Swap Error Contribution (b)')
    ax[1].set_ylabel('Idle Error per N (c)')
    cb2 = fig.colorbar(im2, ax=ax[1])
    cb2.set_label('N for Peak Efficiency')

    # Reference lines and legends
    for i in range(2):
        ax[i].axvline(x=a_fixed, color='black', linestyle='--', alpha=0.7, label=f'a={a_fixed}')
        ax[i].axhline(y=a_fixed, color='black', linestyle='--', alpha=0.7)
        ax[i].legend(loc='upper right')

    plt.tight_layout()
    plt.show()

# Execution loop
a_s = np.array([0.5, 0.1, 0.01, 0.001])
for a in a_s:
    run_parameter_sweep(vareps=1e-10, a_fixed=a, r=2, N_max=100)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def solve_for_distance(vareps, base):
    """
    Calculates D from vareps = base^((D+1)/2)
    Returns D = 2 * log(vareps) / log(base) - 1
    """
    if base >= 1.0 or base <= 0:
        return -1
    d = (2 * np.log(vareps) / np.log(base)) - 1
    return d if d > 0 else -1

def get_efficiency(vareps, a, b, c, N, r=3):
    """
    Calculates efficiency = (r * N * D^2) / ((r + 1) * D'^2)
    """
    D = solve_for_distance(vareps, a)
    
    # Memory base error: base' = a + b + c*N
    base_prime = a + b + c * N
    Dp = solve_for_distance(vareps, base_prime)
    
    if D < 0 or Dp < 0:
        return -1, D, Dp
        
    efficiency = (r * N * (D**2)) / ((r + 1) * (Dp**2))
    return efficiency, D, Dp

def find_optimal_N_stats(vareps, a, b, c, r=3, N_max=1000):
    """
    Finds the N that maximizes efficiency and the range where eff > 1.
    """
    best_eff = -1
    n_opt = None
    n_low, n_high = None, None
    
    # Range of N is limited by the threshold: a + b + cN < 1
    theoretical_max_n = int((1.0 - a - b) / c) if c > 0 else N_max
    search_limit = min(N_max, theoretical_max_n)
    
    for n in range(1, search_limit):
        eff, _, _ = get_efficiency(vareps, a, b, c, n, r)
        
        if eff > 1:
            if n_low is None: n_low = n
            n_high = n
            if eff > best_eff:
                best_eff = eff
                n_opt = n
                
    return best_eff, n_opt, n_low, n_high
def run_parameter_sweep(vareps=1e-10, a_fixed=0.1, r=2, N_max=100):
    b_vals = np.linspace(0.01, 0.5, 100)
    c_vals = np.logspace(-3, 0, 100) 
    B, C = np.meshgrid(b_vals, c_vals)
    
    eff_grid = np.full(B.shape, np.nan)
    n_opt_grid = np.full(B.shape, np.nan)

    for i in range(len(c_vals)):
        for j in range(len(b_vals)):
            eff, n_opt = find_optimal_N_stats(vareps, a_fixed, B[i,j], C[i,j], r, N_max)
            eff_grid[i,j] = eff
            n_opt_grid[i,j] = n_opt

    fig, ax = plt.subplots(1, 2, figsize=(15, 5))
    
    # Efficiency Plot - Using raw strings r'' to avoid SyntaxWarnings
    im1 = ax[0].pcolormesh(B, C, eff_grid, shading='auto', cmap='plasma', vmin=1.0, vmax=10)
    ax[0].set_yscale('log')
    ax[0].set_title(fr'Max Efficiency ($a={a_fixed}$, $\epsilon={vareps}$)')
    ax[0].set_xlabel('Swap Error Contribution (b)')
    ax[0].set_ylabel('Idle Error per N (c)')
    cb1 = fig.colorbar(im1, ax=ax[0], extend='max')
    cb1.set_label('Efficiency Factor')

    # Optimal N Plot
    im2 = ax[1].pcolormesh(B, C, n_opt_grid, shading='auto', cmap='viridis', vmin=1, vmax=N_max)
    ax[1].set_yscale('log')
    ax[1].set_title(fr'Optimal N ($a={a_fixed}$, $N_{{max}}={N_max}$)')
    ax[1].set_xlabel('Swap Error Contribution (b)')
    ax[1].set_ylabel('Idle Error per N (c)')
    cb2 = fig.colorbar(im2, ax=ax[1])
    cb2.set_label('N for Peak Efficiency')

    # Reference lines and legends
    for i in range(2):
        ax[i].axvline(x=a_fixed, color='white', linestyle='--', alpha=0.7, label=f'a={a_fixed}')
        ax[i].axhline(y=a_fixed, color='white', linestyle='--', alpha=0.7)
        ax[i].legend(loc='upper right')

    plt.tight_layout()
    plt.show()

# Execution loop
a_s = np.array([0.1, 0.2, 0.3])
for a in a_s:
    run_parameter_sweep(vareps=1e-10, a_fixed=a, r=2, N_max=100)
```

```python
def p_qubit(t_log_cycle, T_coh_qubit):
    '''
    $$p = \frac{t_{log}}{T_{\text{coh},qub}}$$
    '''
    return t_log_cycle / T_coh_qubit

def p_swap(t_swap, T_coh_cav, T_coh_qubit):
    '''
    2 total swaps, half in qubit, half in cavity
    $$p_{swap} = \frac{t_{swap}}{T_{\text{coh},cav}} + \frac{t_{swap}}{T_{\text{coh},qub}}$$
    '''
    return t_swap / T_coh_cav + t_swap / T_coh_qubit

def p_mem_idle(t_log_cycle, t_swap, N, T_coh_cav):
    '''
    Time spent idle in memory is (N-1) * (t_log_cycle + 2*t_swap) / T_coh_cav
    # '''
    # print(f"Memory idle time: {(N-1) * (t_log_cycle + 2*t_swap)}")
    return (N-1) * (t_log_cycle + 2*t_swap) / T_coh_cav

def find_d_prime(vareps, t_log_cycle, t_swap, N, T_coh_cav, T_coh_qubit, p_q_threshold=0.01,p_mem_threshold = 0.16,
                 return_mem_contrib = False
                 ):
    # $$\varepsilon' = \left(\frac{p + p_{mem}}{p_{\text{threshold}}}\right)^{D'}$$
    
    # print(f"p_qubit: {p_qubit(t_log_cycle, T_coh_qubit)}")
    # print(f"p_mem_idle: {p_mem_idle(t_log_cycle, t_swap, N, T_coh_cav)}")
    # print(f"p_swap: {p_swap(t_swap, T_coh_cav, T_coh_qubit)}")
    p_q =p_qubit(t_log_cycle, T_coh_qubit)
    p_mem =  p_mem_idle(t_log_cycle, t_swap, N, T_coh_cav) + p_swap(t_swap, T_coh_cav, T_coh_qubit)
    base = (p_q/p_q_threshold) + (p_mem/p_mem_threshold)

    mem_contrib_comp_qubit = (p_mem/p_mem_threshold)/(p_q/p_q_threshold)

    if np.round(base,3)>=1: 
        if return_mem_contrib:
            return -1, mem_contrib_comp_qubit
        return -1
    
    if return_mem_contrib:
        return np.log(vareps) / np.log(base), mem_contrib_comp_qubit

    return np.log(vareps) / np.log(base)

def find_d(vareps, t_log_cycle, T_coh_qubit, p_threshold=0.01):
    
    # \varepsilon = \left(\frac{p}{p_{\text{threshold}}}\right)^{D}
    
    p = p_qubit(t_log_cycle, T_coh_qubit)
    base = p/p_threshold
    # print(f"p: {p}, base: {base}")

    if np.round(base,3)>=1: 
        return -1

    return np.log(vareps) / np.log(base)


def find_efficiency(vareps, t_log_cycle, t_swap, T_coh_cav, T_coh_qubit, p_q_threshold=0.01,p_mem_threshold = 0.16, N=5, c= 3):
    '''
    Find how many less control lines we will use with memory 
    '''
    D = find_d(vareps, t_log_cycle, T_coh_qubit, p_q_threshold)
    
    Dp = find_d_prime(vareps, t_log_cycle, t_swap, N, T_coh_cav, T_coh_qubit, p_q_threshold, p_mem_threshold)
    
    resources_with_memory = (c+1)*Dp**2
    resources_without_memory = c*N*D**2
    efficiency = resources_without_memory/resources_with_memory
    return efficiency
# find the lowest N and highest N for which $$2D'^2 <ND^2$$
def find_N_bounds(vareps, t_log_cycle, t_swap, T_coh_cav, T_coh_qubit, p_q_threshold=0.01,p_mem_threshold = 0.16, N_min=1, N_max=500, c= 3):
    '''
    Find the lowest and highest N for which 2D'^2 < N D^2
    Returns (N_low, N_high) or (None, None) if no such N exists.
    '''
    D = find_d(vareps, t_log_cycle, T_coh_qubit, p_q_threshold)
    N_low, N_high = None, None
    N_opt, best_efficiency = None, None
    for N in range(N_min, N_max+1):
        Dp = find_d_prime(vareps, t_log_cycle, t_swap, N, T_coh_cav, T_coh_qubit, p_q_threshold, p_mem_threshold)
        # print(f"N={N}, D'={Dp}, D={D}, 2D'^2={2*Dp**2}, ND^2={N*(D**2)}")
        if Dp < 0 or D < 0:
            continue

        resources_with_memory = (c+1)*Dp**2
        resources_without_memory = c*N*D**2
        efficiency = resources_without_memory/resources_with_memory
        if efficiency>1: # memory is useful
            if N_low is None:
                N_low = N
            N_high = N

            if N_opt is None or efficiency>best_efficiency:
                N_opt = N
                best_efficiency = efficiency
    return N_low, N_high, N_opt, best_efficiency

def sweep_N(vareps, t_log_cycle, t_swap, T_coh_cav, T_coh_qubit, p_q_threshold=0.01,p_mem_threshold = 0.16, N_min=1, N_max=500, c= 3):
    '''
    Find the lowest and highest N for which 2D'^2 < N D^2
    Returns (N_low, N_high) or (None, None) if no such N exists.
    '''
    D = find_d(vareps, t_log_cycle, T_coh_qubit, p_q_threshold)
    efficiencies = []
    mem_contrib_comp_qubits = []
    Ns = []
    ds_ratio_mem_comp_qubits = []
    func = []
    for N in range(N_min, N_max+1):
        # print(f"N={N}")
        # print(find_d_prime(vareps, t_log_cycle, t_swap, N, T_coh_cav, T_coh_qubit, p_q_threshold, p_mem_threshold, return_mem_contrib=True))
        # print(f"N={N}, D'={Dp}, D={D}, 2D'^2={2*Dp**2}, ND^2={N*(D**2)}"))
        (Dp, mem_contrib_comp_qubit) = find_d_prime(vareps, t_log_cycle, t_swap, N, T_coh_cav, T_coh_qubit, p_q_threshold, p_mem_threshold, return_mem_contrib=True)
        # print(f"N={N}, D'={Dp}, D={D}, 2D'^2={2*Dp**2}, ND^2={N*(D**2)}")
        if Dp < 0 or D < 0:
            continue

        resources_with_memory = (c+1)*Dp**2
        resources_without_memory = c*N*D**2
        efficiency = resources_without_memory/resources_with_memory
        efficiencies.append(efficiency)
        mem_contrib_comp_qubits.append(mem_contrib_comp_qubit)
        Ns.append(N)
        ds_ratio_mem_comp_qubits.append(Dp/D)
        # func.append(efficiency -1 
    
    return Ns, efficiencies, mem_contrib_comp_qubits, ds_ratio_mem_comp_qubits

# Example usage:
vareps = 10e-5 # doesnt matter!
t_log_cycle = 1.0e-6 #https://arxiv.org/pdf/2408.13687v1 have 42 ns cz (check readout)
t_swap = 20e-7 #https://arxiv.org/abs/2303.00959 have 100 ns swaps
T_coh_cav = 1e-3
T_coh_qubit = 0.5e-3

# debug d' 
# Dp = find_d_prime(vareps, t_log_cycle, t_swap, 10, T_coh_cav, T_coh_qubit)
# print(f"D': {Dp}")

N_low, N_high, N_opt, optimal_memory_induced_resource_reduction_factor = find_N_bounds(vareps, t_log_cycle, t_swap, T_coh_cav, T_coh_qubit)
print(f"Lowest N: {N_low}, Highest N: {N_high}")
print(f"Optimal N: {N_opt}, Optimal memory-induced resource reduction: {optimal_memory_induced_resource_reduction_factor}")


```

## RAM surface code scaling
$$\varepsilon' = \left(\frac{p + 2Np_{\text{swap}}/8}{p_{\text{threshold}}}\right)^{D'}$$

Control Resources per logical qubit $r = 2D'^2 /N$



```python
def D_prime(vareps_prime, N, p_swap = 0.001, p_threshold = 0.01, p = 0.005):    
    # Solve for D' in terms of vareps_prime and N
    # vareps_prime = ((p + 2*N*p_swap/8) / p_threshold) ** D'
    # Take log on both sides:
    # log(vareps_prime) = D' * log((p + 2*N*p_swap/8) / p_threshold)
    # D' = log(vareps_prime) / log((p + 2*N*p_swap/8) / p_threshold)
    # p = 0.01
    # p_swap = 0.005
    # p_threshold = 0.1
    base = (p + 2 * N * p_swap / 8) / p_threshold

    if base>1: 
        return 100000

    return np.log(vareps_prime) / np.log(base)



def plot_control_resources(vareps_prime_vals, N_vals, D_prime_func):
    import matplotlib.colors as mcolors

    # Compute Z as before
    X, Y = np.meshgrid(1/np.array(vareps_prime_vals), N_vals)
    Z = np.zeros_like(X)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            Dp = D_prime_func(1/X[i, j], Y[i, j])
            if Y[i, j] == 0:
                Z[i, j] = 2 * Dp**2
            else:
                Z[i, j] = 2 * Dp**2 / Y[i, j]

    # Create a masked array for values > 500
    Z_masked = np.ma.masked_where(Z <= 100000, Z)

    plt.figure(figsize=(8, 6))
    c = plt.pcolormesh(X, Y, Z, shading='auto', cmap='viridis', norm=mcolors.Normalize(vmin=Z.min(), vmax=500))
    plt.pcolormesh(X, Y, Z_masked, shading='auto', cmap=mcolors.ListedColormap(['red']), alpha=0.6)
    plt.xscale('log')
    plt.xlabel("Error Correction Gain ($1/\\varepsilon'$)")
    plt.ylabel("Number of Modes (N)")
    plt.title("Control Resources per Logical Qubit ")
    plt.colorbar(c, label="# of lines per logical qubit")
    plt.show()

N_vals = np.arange(0, 20,1)
vareps_prime_vals = np.logspace(-11, -9, 100)
plot_control_resources(vareps_prime_vals, N_vals, D_prime)
# plt.xscale('log')
```

# Now we add different memory sizes 

$$T_q = 119 \mu s$$
$$T_m = 1 \textrm{ms}$$
$$t_{log} = 1.1 \mu s$$
$$t_{swap} = 2 \mu s$$


error on memory $$p_{mem,idle} =  \frac{(N-1) *( t_{log} + 2t_{swap})}{T_{\text{coh},mem}}  $$
error on memory swap $$p_{mem, gate} = t_{swap}*(\frac{1}{T_{\text{coh}, mem}} + \frac{1}{T_{\text{coh}, qub}})$$

$$p_{mem} = p_{mem, idle} + p_{mem, gate}$$

```python
def compute_p_mem(N, t_log=1.1e-6, t_swap=2.639e-6, T_coh_mem = 0.61799e-3, T_coh_qub = 0.119e-3):
    """
    Compute p_mem for different N.
    p_mem = p_mem_idle + p_mem_gate
    p_mem_idle = ((N-1) * (t_log + 2*t_swap)) / T_coh_mem
    p_mem_gate = t_swap * (1/T_coh_mem + 1/T_coh_qub)
    Returns: array of p_mem for each N
    """
    N = np.asarray(N)
    p_mem_idle = ((N - 1) * (t_log + 2 * t_swap)) / T_coh_mem
    p_mem_gate = t_swap * (1 / T_coh_mem + 1 / T_coh_qub)
    return p_mem_idle + p_mem_gate
```

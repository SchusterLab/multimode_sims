# Sweeping Qubit Frequency at fixed jc 

```python
omegas = np.linspace(4, 6, 12)
ej_values = [find_E_J(Ec, omega, is_squid=True) for omega in omegas]
jc_value = 0.5  # uA/um^2
for targ_Ej in ej_values:
    area, pin_length =pin_size_from_Ej(targ_Ej, jc_value, bridge_width=0.2)
    print("--------------------------------------------------")
    print(f"for omega = {np.sqrt(8*Ec*targ_Ej*2)+Ec:.2f} GHz:")
    print(f"targ_Ej: {targ_Ej:.2f} GHz, area: {area:.2f} um^2, pin_length: {pin_length:.2f} um")
    R_n = calculate_R_n(pin_length, bridge_width=0.2, critical_current_per_area=jc_value, is_squid=True)
    print(f"For jc = {jc_value:.2f} nA/um and pin length = {pin_length:.2f} um, expected R_n = {R_n:.2f} Ω")
```

# Sweeping jcs for fixed omega 

```python
omega = 4.5  # GHz
jc_values = np.linspace(0.3, 0.6, 12)  # uA/um^2
for jc_value in jc_values:
    targ_Ej = find_E_J(Ec, omega, is_squid=True)
    area, pin_length =pin_size_from_Ej(targ_Ej, jc_value, bridge_width=0.2)
    print("--------------------------------------------------")
    print(f"for jc = {jc_value:.2f} uA/um^2:")
    print(f"targ_Ej per junction : {targ_Ej:.2f} GHz, area: {area:.2f} um^2, pin_length: {pin_length:.2f} um")
    R_n = calculate_R_n(pin_length, bridge_width=0.2, critical_current_per_area=jc_value, is_squid=True)
    print(f"For jc = {jc_value:.2f} nA/um and pin length = {pin_length:.2f} um, expected R_n for whole squid = {R_n:.2f} Ω")
```
